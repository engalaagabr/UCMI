# CLeaning

### Checking for duplicates

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

### Dealing with outliers

```python
df[(df['year']>2025)]
```

```python
df[(df['year']<1990)]
```

```python
df = df[df["year"].between(1900, 2025)]
```

```python
df=df[(df['selling_price']>0)]
df.describe()
```

```python
df.describe()
```

```python
df[(df['km_driven']>df['km_driven'].mean()+3*df['km_driven'].std())]
```

```python
df.drop([1829],inplace=True)
```

```python
df[(df['km_driven']>df['km_driven'].mean()+3*df['km_driven'].std())]
```

```python
df.drop([3110,3652],inplace=True)
```

### Correcting the datatype of columns

```python
df.info()
```

```python
df['km_driven']=pd.to_numeric(df['km_driven'],errors='coerce')
df.info()
```

### Filling NAN's

```python
df.info()
```

```python
df['km_driven'].mode()
```

```python
df['km_driven'].fillna(df['km_driven'].mode()[0],inplace=True)
```

```python
df['name'].mode()
```

```python
df['name'].fillna(df['name'].mode()[0],inplace=True)
```

```python
df['fuel'].fillna(df['fuel'].mode()[0],inplace=True)
df['owner'].fillna(df['owner'].mode()[0],inplace=True)
df['seller_type'].fillna(df['seller_type'].mode()[0],inplace=True)
df['transmission'].fillna(df['transmission'].mode()[0],inplace=True)
```

```python
df.info()
```

```python
df.describe(include='all')
```

# Final
